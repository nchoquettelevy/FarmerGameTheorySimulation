```python
#Taking minimum 3-month SPEI for dry season (Jan - May)

years = 19 #Because there are 19 years in this dataset

#This creates an empty array that we will populate with the mean temperature for each year, in the for loop below

dry_spei = np.zeros(years)
dry_second = np.zeros(years)
wet_spei = np.zeros(years)
wet_second = np.zeros(years)
total_dry = np.zeros(years)
total_wet = np.zeros(years)

#This for loop goes over each year. We specify the index of the dataset corresponding with the number of the months we are interested in.
#For example, in year 0, we would want the average of the data from index 0 to index 11, which represents the first 12 months of the dataset.
#For the next year, we want to take the average of the data from index 12 to index 23, which represents the next 12 months of the dataset, etc.
for year in range(years):
    
    #Here we calculate the start and end index, based on the year
    start_index = year * 12
    end_index = (year) * 12 + 5
    
    #second_start = year * 12 + 11
    #second_end = (year+1) * 12 + 2
    
    wet_start = year*12 + 6
    wet_end = (year * 12) + 11
    #second_wet_start = year*12 + 9
    #second_wet_end = year*12 + 11

    
    #This takes the average temperature between the start and end indices that were calculated abobve
    dry_spei[year] = recent_spei['spei'][start_index:end_index].min()
    #dry_second[year] = recent_spei['spei'][second_start:second_end].min()
    #total_dry[year] = np.minimum(dry_spei[year], dry_second[year])
    wet_spei[year] = recent_spei['spei'][wet_start:wet_end].max()
    #wet_second[year] = recent_spei['spei'][second_wet_start:second_wet_end].max()
    #total_wet[year] = np.maximum(wet_spei[year], wet_second[year])
    
droughts = np.count_nonzero((dry_spei < -1.0))
floods = np.count_nonzero(wet_spei > 1.0)
print(dry_spei)
print(droughts)
print(wet_spei)
print(floods)

drought_frequency = (droughts) / len(dry_spei)
flood_frequency = floods / len(wet_spei)
print(drought_frequency)
print(flood_frequency)
```
