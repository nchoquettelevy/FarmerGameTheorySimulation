# Importing the Data

```python
#Import Data
raw_cvfs = pd.read_csv('CVFS_Data_001.tsv', sep='\t')

#This command prints out the data "head", i.e. a snapshot of the data columns and the first 5 rows, which is a handy way to examine your data!
raw_cvfs.head()
```

```python
#Selecting Crop Data Only

#This command replaces the double spaces in the original data (which represent N/A or no data) with a 0
raw_cvfs.replace(' ',0, inplace=True)


#This creates a new dataframe to store only the data that we want
maize_cvfs = pd.DataFrame()
maize_cvfs['HHID'] = raw_cvfs['HHID']

rem_cvfs = pd.DataFrame()
rem_cvfs['HHID'] = raw_cvfs['HHID']

crop = 'B'


#This loops over each column name in the raw_cvfs dataframe. If the column name matches one of the variables we want, then we store it in the clean_cvfs dataframe
for item in raw_cvfs.columns:
    
    #For example, here I am saying that any variable that contains 'B11A' should be stored in our dataframe. This includes all the separate 'B11A' variables for each year.
    if crop in item:
        maize_cvfs[item] = raw_cvfs[item].astype(float)
    elif 'REM' in item or 'MIG' in item:
        rem_cvfs[item] = raw_cvfs[item].astype(float)
    
        
rem_cvfs.head()        
maize_cvfs.head()

```

```python
#Dropping Years 63 and 72
cleancrop = pd.DataFrame()
cleancrop = maize_cvfs

#Drop first and last year due to data irregularities
for col in cleancrop.columns:
    #if ('_63' in col) or ('_72' in col):
    if '_63' in col:
        cleancrop.drop(col, axis=1, inplace=True)
        
cleancrop.head()


print(cleancrop.columns[0:40])
```

```python
years = np.arange(64,72,1)
exchange = 73.1 #USD to NPR, average for 2010
#Calculating Income Yields of Cereal Crops
cereal_prices_USD = {'B11': 0.2925, 'B12': 0.289, 'B13': 0.2701, 'B14': 0.8994, 'B15': 1.3902} #Data from FAOSTAT for 2010 USD/kg prices
cereal_prices = {}

for key in cereal_prices_USD:
    cereal_prices[key] = cereal_prices_USD[key] * exchange


cash_crops = ['B17_1', 'B17_2', 'B17_3', 'B17_4', 'B17_5', 'B17_6', 'B17_7']

for col in cleancrop.columns:
    prefix = col[0:5]
    adj_prefix = prefix[0:3]
    if adj_prefix in cereal_prices.keys(): 
        for yr in years:
            cleancrop[adj_prefix + 'Income_' + str(yr)] = cleancrop[adj_prefix + 'C_' + str(yr)] * cereal_prices[adj_prefix]
            cleancrop[adj_prefix + 'Yield_' + str(yr)] = cleancrop[adj_prefix + 'Income_' + str(yr)] / cleancrop[adj_prefix + 'A_' + str(yr)]
            cleancrop[adj_prefix + 'Consumed_' + str(yr)] = (cleancrop[adj_prefix + 'B_' + str(yr)] - cleancrop[adj_prefix + 'C_' + str(yr)]) * cereal_prices[adj_prefix]
            cleancrop[adj_prefix + 'NormConsumed_' + str(yr)] = cleancrop[adj_prefix + 'Consumed_' + str(yr)] / cleancrop[adj_prefix + 'A_' + str(yr)]
    #Calculating Income Yields of Cash Crops
    elif prefix in cash_crops:
        for yr in years:
            cleancrop[prefix + 'Yield_' + str(yr)] = cleancrop[prefix + 'B_' + str(yr)] / cleancrop[prefix + 'A_' + str(yr)]
            
    elif 'B16' in prefix:
        for yr in years:
            cleancrop['B16Yield_' + str(yr)] = cleancrop['B16B_' + str(yr)] / cleancrop['B16A_' + str(yr)]


cleancrop.head()

cleancrop['B15NormConsumed_64'][0:10]
```

```python
#Next Steps - (1) Add Total farm revenue; (2) Drop any columns that do not have yields, HHID, Year

cleancrop.fillna(0, inplace=True)

farm_incomes = pd.DataFrame()

farm_incomes['HHID'] = cleancrop['HHID']

for col in cleancrop.columns:
    if 'Yield' in col:
        farm_incomes[col] = cleancrop[col]
        

farm_incomes
```

```python
#Add total farm incomes for each year
years = np.arange(64,72,1)
crops = ['B11', 'B12', 'B13', 'B14', 'B15', 'B16', 'B17_1', 'B17_2', 'B17_3', 'B17_4', 'B17_5', 'B17_6', 'B17_7']
#crops =  ['B11', 'B12', 'B13', 'B14', 'B15']
avg_size = 0.56 * 29.586 #Average size of farm in kattha

for yr in years:
    farm_incomes['TFR_' + str(yr)] = 0
    for item in crops:
        farm_incomes['TFR_' + str(yr)] += (farm_incomes[item + 'Yield_' + str(yr)] * avg_size / (exchange * 2)) #Include the 2 factor to get avg income per crop cycle

        
farm_incomes
    
    
```

```python
#Select Farm Income only and pivot

income_only = pd.DataFrame()
income_only['HHID'] = farm_incomes['HHID']
for col in farm_incomes.columns:
    if 'TFR' in col:
        income_only[col] = farm_incomes[col]

pivoted_income = pd.wide_to_long(income_only, ['TFR'], i='HHID', j='Year', sep='_').reset_index()

pivoted_income.replace([np.inf, -np.inf], np.nan, inplace=True)
pivoted_income.replace([0], np.nan, inplace=True)
pivoted_income

```

```python
pivoted_income['TFR'].describe()
```

# Calculating Income Correlation

```python
pivoted_incomes = pd.pivot(filtered, index='Year', columns='HHID', values=['TFR'])

clean_incomes = pivoted_incomes.dropna(axis=1)

clean_incomes

#Dropping households that reported no income for entire survey
for i in clean_incomes.columns:
    if clean_incomes[i].sum() == 0:
        clean_incomes.drop([i], axis=1, inplace=True)

clean_incomes.dropna(inplace=True)
uncorr_cols = [218, 406, 438, 486, 507] #These are columns that do not yield a correlation
clean_incomes.drop(clean_incomes.columns[uncorr_cols], axis=1, inplace=True)
clean_incomes
```

```python
#Calculating correlation matrix

matrix = clean_incomes.corr(method='pearson')
running_corr = []

for i in range(len(matrix)):
    for j in range(len(matrix.columns)):
        if i < j:
            running_corr.append(matrix.iloc[i,j])
            if np.isnan(running_corr[-1]) == True:
                print(i, j)

p_avg = np.average(running_corr)
print(p_avg)

matrix

```

```python
#Bootstrapping Correlation Calculation

samples = 1000
households = int(len(matrix) / 151) #151 sample neighbourhoods, total survey sample of 2255
max_hh = len(matrix)
avg_rho = np.zeros(samples)

for a in range(samples):
    random_households = random.sample(range(0, max_hh), households)
    corr_array = []
    
    for i in random_households:
        for j in random_households:
            if i < j:
                corr_array.append(matrix.iloc[i,j])

    avg_rho[a] = np.average(corr_array)


print('Low CI:', np.percentile(avg_rho, 2.5))
print(np.average(avg_rho))
print('High CI:', np.percentile(avg_rho, 97.5))
```

```python
plt.hist(avg_rho)
plt.xlabel('Correlation')
plt.show()
```

# Weibull/Log-Normal Converter

## Weibull

```python
#Weibull Conversion - From Scale (Lambda) and Shape (k) to Mean/Std Dev

lam = 888.207
k = 1.290

mean = lam * math.gamma(1 + 1 / k)
std = math.sqrt(lam ** 2 * (math.gamma(1 + 2 / k) - (math.gamma(1 + 1 / k)) ** 2))

print(mean)
print(std)
```

```python
#Weibull Conversion - From Mean/Std Dev to Shape (k) and Scale (lambda)

mean = 260.9
std_dev = 321.1

k = (std_dev / mean) ** -1.086

lam = (mean / math.gamma(1 + 1 / k))

print('k=',k)
print('lambda=',lam)
```

## Log-Normal

```python
mean = 1016
std_dev = 1083

mu = np.log(mean / math.sqrt(1 + std_dev ** 2 / mean ** 2))

sigma = math.sqrt(np.log(1 + std_dev ** 2 / mean ** 2))
print('mu=',mu)
print('sigma=',sigma)
```

```python
#Log-Normal Conversion - From mu/sigma to mean/std dev of distribution

mu = 5.719
sigma = 1.157

mean = math.exp(mu + sigma ** 2 / 2)

std_dev = math.sqrt((math.exp(sigma ** 2) - 1) * (math.exp(2 * mu + sigma ** 2)))

print("mean:", mean)
print("standard dev:", std_dev)
```
